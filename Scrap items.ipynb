## Scrap Multiple Items using css selector from a single page 



```python
import pandas as pd
```

css 1 : #root > div > div.ant-row.FrEdP.css-1bkhbmc.app > div:nth-child(1) > div > div.ant-col.ant-col-20.ant-col-push-4.Jv5R8.css-1bkhbmc.app > div._17mcb > div:nth-child(2) > div > div > div.ICdUp > div > a

css 2 : #root > div > div.ant-row.FrEdP.css-1bkhbmc.app > div:nth-child(1) > div > div.ant-col.ant-col-20.ant-col-push-4.Jv5R8.css-1bkhbmc.app > div._17mcb > div:nth-child(3) > div > div > div.ICdUp > div > a

css 3 : #root > div > div.ant-row.FrEdP.css-1bkhbmc.app > div:nth-child(1) > div > div.ant-col.ant-col-20.ant-col-push-4.Jv5R8.css-1bkhbmc.app > div._17mcb > div:nth-child(4) > div > div > div.ICdUp > div > a

```python

Collected_data = {
    'Serial Number' : [],
    'Page Address' : [],
    'item number' : [] , 
    'item address' : [] ,
}



for item_number in range(1,41):

    item_num_str = str(item_number)

    CSS_SEL = '#root > div > div.ant-row.FrEdP.css-1bkhbmc.app > div:nth-child(1) > div > div.ant-col.ant-col-20.ant-col-push-4.Jv5R8.css-1bkhbmc.app > div._17mcb > div:nth-child('+item_num_str+') > div > div > div.ICdUp > div > a'

    link = driver.find_element(By.CSS_SELECTOR,CSS_SEL)

    link_extracted = link.get_attribute('href')

    #append as a dictionary 
    Collected_data['Serial Number'].append(item_number)
    Collected_data['Page Address'].append(driver.current_url)
    Collected_data['item number'].append(item_number)
    Collected_data['item address'].append(link_extracted)

    
```

```python
# Convert it as a dataframe 

data = pd.DataFrame(Collected_data)

data.head(12)


```

## Download As csv


```python
data.to_csv("Single_page_multiple_product_dara.csv")
```

## From Multiple page 


```python
# Count page number 


page_text_css = '#root > div > div.ant-row.FrEdP.css-1bkhbmc.app > div:nth-child(1) > div > div.ant-col.ant-col-20.ant-col-push-4.Jv5R8.css-1bkhbmc.app > div.xYcXp > div > div.Ck3Nt > div > div > span:nth-child(1)'

text = driver.find_element(By.CSS_SELECTOR,page_text_css).text

total_page =  math.ceil(int(text.split()[0]) / 40)

print(type(total_page))
```

```python
total_page
```

```python
values = []

for PAGE_NUMBER in range(1,total_page+1):

    link_list = []
    
    driver.get(f' https://www.daraz.com.bd/routers/?page={PAGE_NUMBER}')

    time.sleep(2)

    for item_number in range(1,41):

        item_num_str = str(item_number)

        CSS_SEL = '#root > div > div.ant-row.FrEdP.css-1bkhbmc.app > div:nth-child(1) > div > div.ant-col.ant-col-20.ant-col-push-4.Jv5R8.css-1bkhbmc.app > div._17mcb > div:nth-child('+item_num_str+') > div > div > div.ICdUp > div > a'

        link = driver.find_element(By.CSS_SELECTOR,CSS_SEL)

        link_extracted = link.get_attribute('href')

        link_list.append(link_extracted)

        

    values.append({

        'Page_number' :PAGE_NUMBER ,
        'Items Address' : link_list ,

    })

```

```python
df = pd.DataFrame(values)
```

```python
df.to_csv('multiple_page_data.csv')
```
